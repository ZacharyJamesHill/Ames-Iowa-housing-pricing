## Linear Regression Dropping Low P Values and Cross Validating
***

```python
X = train.drop(columns='saleprice')
y = train['saleprice']

X.drop(columns = high_p_vals, inplace = True)

```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 6)
```

```python
lr = LinearRegression()

print(cross_val_score(lr, X_train, y_train, cv=5).mean()
)

lr.fit(X_train, y_train)

lr.score(X_train, y_train), lr.score(X_test, y_test)

```

```python
testing = test.drop(columns = 'saleprice')

testing.drop(columns = high_p_vals, inplace = True)


output = pd.DataFrame({'Id': test['id']})


output['SalePrice'] = lr.predict(testing)


output['Id'] = [int(val) for val in output['Id']]


pd.DataFrame.to_csv(output, '../../submissions/second_lr.csv', index = False)
```
